```python
import pandas as pd

df = pd.read_csv("data/Ivanovskoe.csv")
df
```

```python
from GeneratorModel.DayConsumption import DayWaterConsumptionModel

pd.DataFrame(
    DayWaterConsumptionModel.day_consumption_model.generate(
        df.iloc[:, 1:],
        1, 
        "DATE_WEEK"
    )
)
```

```python
df_new = pd.DataFrame(
    df.groupby("УНОМ").first()
).reset_index()
```

```python
df_new
```

```python
from GeneratorModel import pipeline_generator
from datetime import datetime, timedelta


res = pipeline_generator.generate_pipeline_for_date_to_date(
    df_new.iloc[:, 1:], 
    "DATE_WEEK",
    datetime(2025, 1, 1), 
    datetime(2025, 12, 31), 
)
```

```python
res_reshaped = res.transpose(1, 0, 2).reshape(res.shape[1], -1)
```

```python
cols = []
begin_date = datetime(2025, 1, 1)
end_date = datetime(2025, 12, 31)
current_date = begin_date

ENG_DATE_TO_RUS = {
    "Monday": "Понедельник",
    "Tuesday": "Вторник",
    "Wednesday": "Среда",
    "Thursday": "Четверг",
    "Friday": "Пятница",
    "Saturday": "Суббота",
    "Sunday": "Воскресенье",
}


while current_date <= end_date:
    for i in range(24):
        cols.append(f"{str(current_date).split()[0]}_{ENG_DATE_TO_RUS[current_date.strftime('%A')]}_{i + 1}")
    current_date += timedelta(days=1)
```

```python
info = pd.DataFrame(res_reshaped, columns=cols)
```

```python
info["UNOM"] = df_new["УНОМ"]
```
